```python
import pandas as pd
```

```python
df=pd.read_csv("Student_Performance.csv")
df.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df)
```

```python
df.shape
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Previous Scores', 'Hours Studied', 'Sleep Hours', 'Sample Question Papers Practiced']]
y=df[['Performance Index']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8, random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred[:5]
```

# WE EVALUATE OUR MODEL HERE

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
accurate = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R square is {accurate}")    #  One you have one here, it is correct but it is over fitting and MSE must be 0
print(f"MSE 1 is {mse}")
```

## VISUALIZE THE EVALUATION PREDICTED

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel("Previous Scores, Hours Studied, Sleep Hours and Sample Question Papers Practiced")
plt.ylabel("Performance Index")
plt.title("Previous Scores, Hours Studied, Sleep Hours and Sample Question Papers Practiced vs Predicted Performance Index")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')  # Diagonal line
plt.show()
```
